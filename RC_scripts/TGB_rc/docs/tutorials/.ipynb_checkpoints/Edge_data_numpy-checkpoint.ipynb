# Access edge data as numpy arrays

This tutorial will show you how to access various datasets and their corresponding edgelists in `tgb`

You can directly retrieve the edge data as `numpy` arrays, `PyG` and `Pytorch` dependencies are not necessary

The logic is implemented in `dataset.py` under `tgb/linkproppred/` and `tgb/nodeproppred/` folders respectively


```python
from tgb.linkproppred.dataset import LinkPropPredDataset
```

specifying the name of the dataset

```python
name = "tgbl-wiki" 
```

### process and loading the dataset

if the dataset has been processed, it will be loaded from disc for fast access

if the dataset has not been downloaded, it will be processed automatically

```python
dataset = LinkPropPredDataset(name=name, root="./Datasets", preprocess=True)
type(dataset)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Convert the series to a DataFrame for easier manipulation
dest_counts_df = dest_counts_per_source.reset_index(name='UniqueDestCount')

# Calculate the CDF
dest_counts_df = dest_counts_df.sort_values(by='UniqueDestCount')
dest_counts_df['cdf'] = dest_counts_df['UniqueDestCount'].rank(method='average', pct=True)

# Plotting the CDF
plt.figure(figsize=(8, 6))
plt.plot(dest_counts_df['UniqueDestCount'], dest_counts_df['cdf'], marker='.', linestyle='none')

plt.xscale('log')  # Set the x-axis to a logarithmic scale

plt.title('CDF of Number of Unique Destination Nodes per Source Node')
plt.xlabel('Number of Unique Destination Nodes')
plt.ylabel('CDF')
plt.grid(True)

plt.show()
```

```python
# Assuming dest_counts_per_source is a Series with the count of unique destinations for each source
percentiles = [0.25, 0.5, 0.75, 0.9, 0.99]
percentile_values = dest_counts_per_source.quantile(percentiles)

print("Percentiles for the Number of Unique Destination Nodes per Source Node:")
print(percentile_values)

# Additional insights
mean = dest_counts_per_source.mean()
std_dev = dest_counts_per_source.std()
max_value = dest_counts_per_source.max()
min_value = dest_counts_per_source.min()

print(f"Mean: {mean}, Std Dev: {std_dev}, Max: {max_value}, Min: {min_value}")
```

```python
# Define your subset of source nodes
source_subset = np.arange(1000,1300, 1)# Example subset

# Filter the DataFrame for these source nodes
filtered_data = data_df[data_df['Source'].isin(source_subset)].copy()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Ensure Timestamp is in datetime format
filtered_data['Timestamp'] = pd.to_datetime(filtered_data['Timestamp'])

# Sort data by timestamp
filtered_data.sort_values(by='Timestamp', inplace=True)

# Create a figure and axis for the plot
fig, ax = plt.subplots(figsize=(12, 8))

# Generate a color for each source node in the filtered subset
colors = plt.cm.jet(np.linspace(0, 1, len(source_subset)))

# Map each source node to a color
color_map = {source: color for source, color in zip(source_subset, colors)}

# Plot each edge, coloring by source node
for (source, destination), group in filtered_data.groupby(['Source', 'Destination']):
    ax.scatter(group['Timestamp'], [destination] * len(group), color=color_map[source], label=source, alpha=0.5)

# Create a custom legend
from matplotlib.lines import Line2D
legend_elements = [Line2D([0], [0], marker='o', color='w', label=source,
                          markerfacecolor=color, markersize=10)
                   for source, color in color_map.items()]
ax.legend(handles=legend_elements, title="Source Nodes", loc='best')

# Improve readability of the plot
ax.set_xlabel('Time')
ax.set_ylabel('Destination Node')
ax.set_title('Edges Over Time by Source Node')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_metrics(data_df):
    # Convert 'Timestamp' to datetime if not already
    data_df['Timestamp'] = pd.to_datetime(data_df['Timestamp'])

    # Get the range of timestamps
    min_time = data_df['Timestamp'].min()
    max_time = data_df['Timestamp'].max()

    # Generate 100 time bins
    time_bins = pd.date_range(start=min_time, end=max_time, periods=101)

    # Metrics to store
    avg_clustering = []
    avg_path_length = []
    graph_density = []

    for i in range(len(time_bins) - 1):
        # Filter edges within the current time bin
        bin_edges = data_df[(data_df['Timestamp'] >= time_bins[i]) & (data_df['Timestamp'] < time_bins[i+1])]
        
        # Create a graph from the bin_edges
        G = nx.from_pandas_edgelist(bin_edges, 'Source', 'Destination', create_using=nx.Graph())
        
        # Calculate metrics
        # Clustering
        clustering = nx.average_clustering(G)
        avg_clustering.append(clustering)
        
        # Average path length - skip if graph is not connected
        if nx.is_connected(G):
            path_length = nx.average_shortest_path_length(G)
        else:
            path_length = np.nan
        avg_path_length.append(path_length)
        
        # Density
        density = nx.density(G)
        graph_density.append(density)
    
    # Plotting
    fig, ax = plt.subplots(3, 1, figsize=(10, 15), sharex=True)
    
    # Time bins for x-axis (use the midpoint of each bin for plotting)
    x = [time_bins[i] + (time_bins[i+1] - time_bins[i]) / 2 for i in range(len(time_bins) - 1)]
    
    ax[0].plot(x, avg_clustering, label='Average Clustering Coefficient')
    ax[0].set_ylabel('Clustering Coefficient')
    ax[0].legend()
    
    ax[1].plot(x, avg_path_length, label='Average Path Length', color='orange')
    ax[1].set_ylabel('Path Length')
    ax[1].legend()
    
    ax[2].plot(x, graph_density, label='Graph Density', color='green')
    ax[2].set_ylabel('Density')
    ax[2].set_xlabel('Time')
    ax[2].legend()
    
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Assuming data_df is your DataFrame
# calculate_and_plot_metrics(data_df)
```

```python
calculate_and_plot_metrics(data_df)
```

```python
from astropy.stats import LombScargle
import numpy as np
import matplotlib.pyplot as plt

def analyze_periodicity_with_lomb_scargle(time_series):
    """
    Analyze and plot the periodicity of an irregular time series using the Lomb-Scargle Periodogram.

    Parameters:
    - time_series: DataFrame with columns 'Timestamp' and 'Value', where 'Timestamp' is the time of each observation,
      and 'Value' is the measurement or count of events.
    """
    # Convert timestamps to a numerical format (e.g., seconds since start)
    timestamps = (time_series['Timestamp'] - time_series['Timestamp'].min()).dt.total_seconds()
    values = time_series['Value']
    
    # Calculate the Lomb-Scargle periodogram
    frequency, power = LombScargle(timestamps, values).autopower()
    
    # Plot the periodogram
    plt.figure(figsize=(10, 6))
    plt.plot(frequency, power)
    plt.title('Lomb-Scargle Periodogram')
    plt.xlabel('Frequency')
    plt.ylabel('Power')
    plt.show()
    
    # Identifying the dominant frequency
    idx_max_power = np.argmax(power)
    dominant_frequency = frequency[idx_max_power]
    print(f"Dominant Frequency: {dominant_frequency} with power {power[idx_max_power]}")

# Example usage
# Assuming you have a DataFrame for a specific source-destination pair, with 'Timestamp' and corresponding values (e.g., interaction counts)
# analyze_periodicity_with_lomb_scargle(your_time_series_dataframe)
```
